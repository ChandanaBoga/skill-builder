```python
#  1.Find the sum of all numbers using reduce()

from functools import reduce
l=[1,2,3,4]
y=reduce(lambda a,b:a+b,l)
print(y)
```

```python
#  2.Find the product of all numbers using reduce()

from functools import reduce
l=[1,2,3,4]
y=reduce(lambda a,b:a*b,l)
print(y)
```

```python
#  3. Find the maximum number using reduce()

from functools import reduce
l=[1,2,3,4,5,2]
y=reduce(lambda a,b:a if a>b else b,l)
print(y)
```

```python
#  5. Find the maximum length word using reduce()


from functools import reduce
l=["apple","bat","ca","doggyee","app"]
y=reduce(lambda a,b:a if len(a)<len(b) else b,l)
print(y)
```

```python
#  6. Find factorial of a number using reduce()

def factorial(x,y):
    return x*y
from functools import reduce
n=5
y=reduce(factorial,range(1,n+1))
print(y)
```

```python
#  8.  Find the count of even numbers using reduce()

def even(acc,x):
    if x%2==0:
        return acc+1
    return acc
l=[1,2,3,4,5,6]
from functools import reduce
y=reduce(even,l,0)
print(y)
```

```python
#  9.  Find the count of odd numbers using reduce()

def odd(acc,n):
    if n%2==1:
        return acc+1
    return acc
l=[1,2,3,4,5,6]
from functools import reduce
y=reduce(odd,l,0)
print(y)
```

```python
#  10.  Find the sum of squares using reduce()

l=[2,3,1]
from functools import reduce
y=reduce(lambda x,y:x+(y**2),l,0)
print(y)
```
