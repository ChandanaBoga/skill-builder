```python
l=[5,10,15,20,25,30]
x=list(map(lambda a:a**2,l))
y=list(filter(lambda a:True if a%5==0 else False,x))
z=sorted(y,reverse=True)
from functools import reduce
a=reduce(lambda a,b:a+b,z)
print(a)
final=reduce(lambda a,b:a+b,sorted(list(filter(lambda a:True if a%5==0 else False,list(map(lambda a:a**2,l)))),reverse=True))
print(final)
```
