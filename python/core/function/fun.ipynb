```python
def multiply(*args):
    mul=1
    for i in range (0,len(args)):
        mul*=args[i]
    return mul
print(multiply(1,2,3,4))
```

```python
#   6.	Create a function apply(func, value) that takes another function as argument. Pass a function that returns cube of a number.
def cube(n):
    return n*n*n
def apply(func, value):
    return func(value)
print(apply(cube,3))
```

```python
#  7.	Create a function calculator(a, b, operation) and pass:
#  •	add function
#  •	subtract function
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def calculator(a,b,operation):
    return operation(a,b)
print(calculator(2,3,add))
print(calculator(4,3,sub))

```

```python
#  8.	Create a function is_even(num) and pass it into another function called check().


def is_even(num):
    return num%2==0

def check(fun,n):
    if fun(n):
        print("even")
    else:
        print("odd")
check(is_even,5)
```
