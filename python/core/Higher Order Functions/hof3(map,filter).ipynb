```python
#  1.	Use map() with a lambda function to add 5 to every element.

l=[1,2,3,4]
x=list(map(lambda a:a+5,l))
print(x)
```

```python
l=[[1,2],[3,4],[5,6]]
x=list(map(lambda x:x+[5],l))
print(x)
```

```python
#  2.	Use map() to convert all elements of a list to their squares.

l=[1,2,3,4]
y=list(map(lambda x:x**2,l))
print(y)
```

```python
#  3.	Use map() to find the cube of each number.

l=[1,2,3,4]
y=list(map(lambda x:x**3,l))
print(y)
```

```python
# 4.	Use map() to convert a list of integers into strings.

l=[1,2,3,4]
y=list(map(str,l))
print(y)
```

```python
#  5.	Use map() to find the length of each string.

l=["abs","abbb","a"]
y=list(map(lambda x:len(x),l))
print(y)
```

```python
l=["abs","abbb","a"]
y=list(map(len,l))
print(y)
```

```python
def length(x):
    return len(x)
l=["abs","abbb","a"]
y=list(map(length,l))
print(y)
```

```python
#  6.	Convert temperatures from Celsius to Fahrenheit using map().
#  •	Formula: F = (C * 9/5) + 32

def fah(C):
    F=(C*9/5)+32
    return F
l=[23,56,45]
y=list(map(fah,l))
print(y)
```

```python
#  7.	Use map() to check whether numbers are even or odd.

def num(n):
    if n%2==0:
        print("even")
    else:
        print("odd")
l=[1,2,3,4,5,6]
y=list(map(num,l))
```

```python
def num(n):
    if n%2==0:
        return "even"
    else:
        return "odd"
l=[1,2,3,4,5,6]
y=list(map(num,l))
print(y)
```

```python
# 8.	Filter only even numbers from a list.
def even(n):
    return n%2==0
l=[1,2,3,5,7,9,10]
y=list(filter(even,l))
print(y)
```

```python
# 9.	Filter only odd numbers from a list.

def odd(n):
    return n%2==1
l=[1,2,3,5,7,9,10]
y=list(filter(odd,l))
print(y)
```

```python
# 10.	 Filter numbers greater than 10 from a list.

l=[1,3,10,23,45,2,23]
y=list(filter(lambda x:x if x>10 else False,l))
print(y)
```

```python
#  11.	 Filter strings with length greater than 4 from a list.

l=["apple","bat","cat","doggy0"]
y=list(filter(lambda x:x if len(x)>4 else False,l))
print(y)
```

```python
#  12.	 Filter numbers divisible by 5 from a list.

l=[10,20,30,22,33,44,25]
y=list(filter(lambda x:x if x%5==0 else False,l))
print(y)
```
