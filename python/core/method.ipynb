```python
class A:
    def __init__(self,q,r):
        self.q=q
        self.r=r
    @classmethod
    def divideby10(cls,n):
        return cls(n//10,n%10)
obj=A.divideby10(25)
print(obj.q)
print(obj.r)
```

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def is_Passed(self):
        if self.marks>40:
            print("pass")
        else:
            print("fail")

s1=Student("ram",80)
print(s1.name)
print(s1.marks)
s2=Student("abc",30)
print(s2.name)
print(s2.marks)
s1.is_Passed()
s2.is_Passed()
print(s1.is_Passed())




```

```python
class Employee:
    company_name="TechCorp"
    def __init__(self,name):
        self.name=name
    @classmethod
    def change_company(cls,new_name):
        cls.company_name=new_name
e1=Employee("ABC")
e2=Employee("XYZ")
print(e1.company_name)
print(e2.company_name)
Employee.change_company("corp")
print(e1.company_name)
```

```python
class Car:
    wheels=4
    def __init__(self,mileges):
        self.mileges=mileges
    def display_specs(self):
        print("mileges:",self.mileges)
        print("wheels:",Car.wheels)
    @classmethod
    def change_wheels(cls,new_wheels):
        cls.wheels=new_wheels
c1=Car(20)
c1.display_specs()
Car.change_wheels(6)
c1.display_specs()



```

```python
class Temperature:
    def __init__(self,celsius):
        self.celsius=celsius
    @staticmethod
    def to_fahrenheit(n):
        return(n*9/5)+32
    def show_conversion(self):
        f=Temperature.to_fahrenheit(self.celsius)
        print(self.celsius)
        print(f)
t1=Temperature(25)
t1.show_conversion()


```
