```python
def is_prime(n):
    fc=0
    for i in range(1,n+1):
        if n%i==0:
            fc+=1
    return fc==2
n=int(input())
i=1
count=0
while count<n:
    b=is_prime(i)
    if b==True:
        print(i)
        count+=1
    i+=1
```
