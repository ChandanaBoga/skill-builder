```python
#  1.	Write a lambda function that takes a number and returns its cube.

cube=lambda a:a**3
print(cube(3))
```

```python
#  2.	Create a lambda that takes two numbers and returns the larger one using a conditional expression (x if x > y else y).

largest=lambda x,y:x if x>y else y
print(largest(2,3))
```

```python
#  3.	Convert this regular function into a lambda: def even(n): return n % 2 == 0

even= lambda n: n%2==0
print(even(4))
```

```python
#  4.	Write a lambda function to find the square of a number.

sq=lambda a:a**2
print(sq(3))
```

```python
#  5.	Write a lambda function to find the maximum among three numbers.

x=lambda a,b,c:a if a>b and a>c  else (b if b>c else c)
print(x(1,2,3))
```

```python
# 6.	Use lambda to create a function that returns:
# •	"Positive" if number > 0
# •	"Negative" if number < 0
# •	"Zero" otherwise

n= lambda n:'positive' if n>0 else('negative' if n<0 else 'zero')
print(n(7))
print(n(-7))
print(n(0))



```
