```python
# 1.	Write a function intro(name, city, hobby) that prints a sentence about a person. Call it in two different orders and observe the difference.

def intro(name,city,hobby):
    print("my name is",name,"i am from",city,"and my hobby is",hobby)
intro("chandana","hyd","reading")
intro("reading","hyd","chandana")


```

```python
#  2.	Create subtract(a, b) that returns a - b. What is the difference between subtract(10, 3) and subtract(3, 10)?

def subtract(a,b):
    return a-b
print(subtract(10,3))
print(subtract(3,10))
```

```python
# 3.	Write a function bio(first_name, last_name, age) and call it correctly using positional arguments.


def bio(first_name,last_name,age):
    print(first_name,last_name,age)
bio("chandana","boga",20)
bio(20,"chandana","boga")
```

```python
# 4.	Call the function send_email(to, subject, body) using keyword arguments in any order.


def send_email(to,subject,body):
    print(to,subject,body)
send_email(to="chandana",subject="leave request",body="requesting leave for 4 days")
send_email(subject="leave request",body="requesting leave for 4 days",to="chandana")
```

```python
# 5.	Write a function create_profile(username, email, age) and call it using keyword arguments.

def create_profile(username,email,age):
    print("username",username,"email",email,"age",age)
create_profile(username="abc",email="abc@gmail.com",age=20)
create_profile(age=20,username="abc",email="abc@gmail.com")
```

```python
# 6.	Write a function power(base, exponent=2) that returns base^exponent. Test with one and two arguments.

def power(base, exponent=2):
    return base**exponent
print(power(2,2))
print(power(3))
```

```python
# 7.	Create a function connect(host, port=3306, protocol='TCP') and call it with various combinations.

def connect(host,port=3306,protocol="tcp"):
    print(host,port,protocol)
connect("abc")
connect("abc",3000,"udp")
connect("abc","udp",3000)
```

```python
# 8.	Write a function discount_price(price, discount=10) that returns the discounted price. Test with and without the discount argument.

def discount_price(price,discount=10):
    return price-(discount/100*price)
print(discount_price(100))
```
