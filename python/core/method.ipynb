```python
class Temperature:
    def __init__(self,celsius):
        self.celsius=celsius
    @staticmethod
    def to_fahrenheit(n):
        return(n*9/5)+32
    def show_conversion(self):
        f=Temperature.to_fahrenheit(self.celsius)
        print(self.celsius)
        print(f)
t1=Temperature(25)
t1.show_conversion()


```

```python
class Book:
    total_books=0
    def __init__(self,title,author):
        self.title=title
        self.author=author
        Book.total_books+=1
    @classmethod
    def from_string(cls,book_str):
        title,author=book_str.split("-")
        return cls(title,author)
    @staticmethod
    def is_valid_title(title):
        if len(title)>=3:
            return True
        return False
title1="python"
author1="guido"
if Book.is_valid_title(title1):
    b1=Book(title1,author1)
    print(b1.title, "-", b1.author)
x="java-james"
title2=x.split("-")[0]
if Book.is_valid_title(title2):
    b2=Book.from_string(x)
    print(b2.title, "-", b2.author)
```

```python
class Employee:
    bonus_rate=0.1
    def __init__(self,name,base_salary):
        self.name=name
        self.base_salary=base_salary
    def final_salary(self):
        return self.base_salary+(self.base_salary*Employee.bonus_rate)
    @classmethod
    def update_bonus(cls,new_rate):
        cls.bonus_rate=new_rate
    @staticmethod
    def is_valid_salary(sal):
        if sal>0:
            return True
        else:
            return False
if Employee.is_valid_salary(5000):
    e1=Employee("ABC",5000)
    print(e1.final_salary())
if Employee.is_valid_salary(6000):
    e2=Employee("XYZ",6000)
    print(e2.final_salary())
e1.update_bonus(0.2)
print(e1.final_salary())
print(e2.final_salary())
Employee.update_bonus(0.2)
print(e1.final_salary())
print(e2.final_salary())

```

```python
class BankAccount:
    bank_name="SBI"
    def __init__(self,holder,balance):
        self.holder=holder
        self.balance=balance
    def deposit(self,amount):
        if self.validate_amount(amount):
            self.balance=self.balance+amount
            print(self.holder,amount)
            print(self.balance)
    @classmethod
    def change_bank_name(cls,new_name):
        cls.bank_name=new_name
    @staticmethod
    def validate_amount(amount):
        return amount>0
a1=BankAccount("abc",1000)
print(BankAccount.bank_name)
print(a1.holder)
print(a1.balance)
a1.deposit(100)
BankAccount.change_bank_name("KOTAK")
print(BankAccount.bank_name)

```

```python
class Student:
    passing_marks=40
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def result(self):
        if self.marks>=Student.passing_marks:
            print(self.name,"pass")
        else:
            print(self.name,"fail")
    @classmethod
    def update_passing_marks(cls,new_marks):
        cls.passing_marks=new_marks
    @staticmethod
    def grade_category(marks):
        if marks>=90 and marks<=100:
            return "A"
        elif marks>=80 and marks<=89:
            return "B"
        else:
            return "C"
s1=Student("abc",95)
print(s1.name)
print(Student.grade_category(s1.marks))
s1.result()
```
