```python
l=[10, 350, 10, 350, 20]
y=list(map(id,l))
print(l)
print(y)


# ✅ Example Output (Your addresses will be different)
# Numbers: [10, 350, 10, 350, 20]
# Memory addresses: [140734..., 140735..., 140734..., 140735..., 140736...]
# You will notice:
# Address of first 10 = Address of second 10
# Address of first 350 = Address of second 350
# Some addresses repeat.

# ✅ Why Do Some Addresses Repeat?
# This happens because of object reuse (interning) in Python.
# 🔹 1. Small Integer Caching:Python automatically stores small integers (usually from -5 to 256) in memory.
# 10 is stored once and reused everywhere.
# That’s why both 10s have the same memory address.
# 🔹 2. Same Value = Same Object (Sometimes)
# If the same integer value appears multiple times in a list, Python may reuse the same object instead of creating a new one.
# So both 350 values may also point to the same memory location.

# 🔎 What id() Does
# id() returns the memory address of an object.
# If two variables have the same id(), they refer to the same object in memory.

# | Value | Why Address Repeats            |
# | ----- | ------------------------------ |
# | 10    | Small integer cached by Python |
# | 350   | Python reused the same object  |
# | 20    | Small integer cached           |

# Python avoids creating duplicate immutable objects (like integers) to save memory.
# This is called:
# Object Interning / Caching
```

```python
# list(map(str, [1, 2, 3]))
# str(1)
# str(2)
# str(3)
# output:['1', '2', '3'](more better as it is built in function str is a built-in C-implemented function, No extra lambda wrapper, Fewer  function calls, Less overhead)


# list(map(lambda x: str(x), [1, 2, 3]))
# (lambda x: str(x))(1)
# (lambda x: str(x))(2)
# (lambda x: str(x))(3)
# output:['1', '2', '3']


# | Feature                | `map(str, ...)` | `map(lambda x: str(x), ...)` |
# | ---------------------- | --------------- | ---------------------------- |
# | Uses built-in function | ✅ Yes           | ❌ No (wraps inside lambda)   |
# | Extra function call    | ❌ No            | ✅ Yes                        |
# | Simpler                | ✅ Yes           | ❌ More complex               |
# | Performance            | 🚀 Faster       | Slightly slower              |

# 🚀 Which One is Faster?
# ✅ map(str, [1, 2, 3]) is faster
# 👉 Why?
# str is a built-in C-implemented function
# No extra lambda wrapper
# Fewer function calls
# Less overhead

# The lambda version adds:
# Extra function layer
# Extra call for every element
# So it is slightly slower.
```

```python
# 10. Given a list of numbers:
# [5, 10, 15, 20, 25, 30]
# Perform the following in a single pipeline:
# • Use map() to square each number
# • Use filter() to keep only numbers divisible by 5
# • Use reduce() to calculate the sum of remaining numbers

from functools import reduce
l=[5, 10, 15, 20, 25, 30]
sq=list(map(lambda x:x**2,l))
print(sq)
fil=list(filter(lambda x: x%5==0,sq))
print(fil)
red=reduce(lambda x,y:x+y,fil)
print(red)

red=reduce(lambda x,y:x+y,filter(lambda x: x%5==0,map(lambda x:x**2,l)))
print(red)




```
