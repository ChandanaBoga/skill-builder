```python
def cube(n):
    return n**3
print(cube(3))
```

```python
def is_even(n):
    if n%2==0:
        return True
    return False
print(is_even(3))
print(is_even(2))
```

```python
def full_name(first,last):
    return first + last
print(full_name("boga","chandana"))
```

```python
def area_of_rectangle(length,width):
    return length*width
print(area_of_rectangle(6,4))
```
