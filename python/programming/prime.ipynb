```python
# nearest prime
def is_prime(n):
    fc=0
    for i in range(1,n+1):
        if n%i==0:
            fc+=1
    if fc==2:
        return True
    return False
n=int(input())
i=1
bp=0
while True:
    m=n+i
    if is_prime(m):
        bp=m
        break
    i+=1
ap=0
while True:
    m=n-i
    if is_prime(m):
        ap=m
        break
    i-=1
bpc=n-bp
apc=ap-n
if bpc==apc:
    print(bp,ap)
elif bpc>apc:
    print(ap)
elif apc>bpc:
    print(bp)
```
