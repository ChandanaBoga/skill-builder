```python
#Q5. Explain in your own words: why do we use functions instead of writing code directly?

#Reasons for using functions:
#Code Reusability – A function can be called many times without rewriting the same code.
#Reduces Repetition – It avoids duplicate code and makes programs shorter.
#Improves Readability – Functions divide a large program into smaller, meaningful parts.
#Easy Debugging and Maintenance – Errors can be fixed in one place instead of multiple places.
#Modularity – Different tasks can be handled by different functions, making teamwork easier.
```
