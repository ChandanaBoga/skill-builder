```python
#  Q1. Call the function send_email(to, subject, body) using keyword arguments in any order.

def send_email(to,subject,body):
    print("to:",to)
    print("subject:",subject)
    print(body)
send_email(to="xyz@gmail.com",subject="leave letter",body="requesting leave for 3 days")

```

```python
#  Q2. Write a function create_profile(username, email, age) and call it using keyword arguments.

def create_profile(username,email,age):
    print("username:",username)
    print("email:",email)
    print("age:",age)
create_profile(username="abc",email="abc@gmail.com",age=21)
```

```python
#  Q4. Rewrite this call using keyword arguments: book_ticket('Alice', 'Delhi', 'Mumbai', 2)

def book_ticket(name,from_dest,to,tickets):
    print("name:",name)
    print("from:",from_dest)
    print("to:",to)
    print("no of tickects:",tickets)
book_ticket(name="alice",from_dest="delhi",to="mumbai",tickets=2)
```
