```python
#  example

def greet():
    print("good morning")

def my_decor(func):
    def wrapper():
        print("hello")
        func()
        print("how are you?")
    return wrapper

a=my_decor(greet)
a()
```

```python
#  @ decorator syntax

def my_decor(func):
    def wrapper():
        print("hello")
        func()
        print("how are you?")
    return wrapper

@my_decor
def greet():
    print("good morning")
greet()

```

```python
#  function with parameter

def decor1(func):
    def wrapper(a,b):
        print("my bio is")
        func(a,b)
    return wrapper

@decor1
def bio(name, age):
    print(f"my name is {name}, age is {age}")

# a=decor1(bio)
# a("chandana",21)
bio("chandana",21)
```

```python
#  1. Create a decorator that prints a message before and after executing a function.
	#Expected output:
	#Before function execution
	#Hello Student
	#After function execution

def decor(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper
@decor
def greet():
    print("Hello Student")
greet()
```

```python
#2. Create a decorator my_decorator for the below function:
	#def greet():
    		#print("Good Morning")
# Output should be:
#Welcome Message
#Good Morning
#Thank You Message


def my_decorator(func):
    def wrapper():
        print("Welcome Message")
        func()
        print("Thank You Message")
    return wrapper
@my_decorator
def greet():
    print("Good Morning")
greet()
```

```python
#  3. Create a decorator that accepts a function with parameters.
#Example:
#@calculate
#def add(a,b):
 #   print(a+b)

#add(10,20)
#Output:
#Addition result: 30


def calculate(func):
    def wrapper(a,b):
        print("Addition result: ",end="")
        func(a,b)
    return wrapper
@calculate
def add(a,b):
    print(a+b)

add(10,20)
```

```python
#4. Create a decorator to check whether a number is even or odd before executing a function.
#Example:
#@check_number
#def show(n):
 #   print("Number accepted")
#Input:
#show(10)
#Output:
#Even number
#Number accepted

def check_number(func):
    def wrapper(n):
        if n%2==0:
            print("Even NUmber")
            func(n)
        else:
            print("Odd number")
    return wrapper

@check_number
def show(n):
    print("Number accepted")
show(10)
```

```python
#5. Create a login verification decorator.
#Requirement:
#If user is logged in:
#	Access granted
#	Welcome to profile
#Otherwise:
#	Please login first
#Example:

login=True
def login_required(func):
    def wrapper():
        if login:
            print("Access granted")
            func()
        else:
            print("Please login first")
    return wrapper


@login_required
def profile():
    print("Welcome to profile")
profile()
```
