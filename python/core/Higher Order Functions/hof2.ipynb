```python
# 1. Explain the difference between map(), filter(), and reduce() in Python.
# • What does each function return?
# • When should you prefer lambda functions over normal functions?


# | Function   | Purpose                             | What It Returns                |
# | ---------- | ----------------------------------- | ------------------------------ |
# | `map()`    | Transforms each element             | A **map object** (iterator)    |
# | `filter()` | Filters elements based on condition | A **filter object** (iterator) |
# | `reduce()` | Reduces list to single value        | A **single value**             |


# 1️⃣ map()
# ✅ Purpose:Applies a function to every element in an iterable.
# example:
# numbers = [1, 2, 3]
# result = list(map(lambda x: x * 2, numbers))
# print(result)
# output:[2, 4, 6]


# 2️⃣ filter()
# ✅ Purpose:Keeps elements that satisfy a condition.
# Example:
# numbers = [1, 2, 3, 4]
# result = list(filter(lambda x: x % 2 == 0, numbers))
# print(result)
# Output:[2, 4]


# 3️⃣ reduce()
# (From functools module)
# ✅Purpose:Reduces list into a single value.
# Example:
# from functools import reduce
# numbers = [1, 2, 3, 4]
# result = reduce(lambda a, b: a + b, numbers)
# print(result)
# Output:10


# ✅ When Should You Prefer Lambda Over Normal Function?
# ✔ Use lambda when:
# The function is short (1 line)
# Used only once
# Simple logic
# Inside map(), filter(), reduce()

```

```python
# 4. Given a list:
# nums = [1, 2, 3, 4]
# Use reduce() with a lambda to compute the sum, but start with an initial value
# of 10.
# Explain how the initial value affects the reduction process.



# ✅ Code Using reduce() with Initial Value
# from functools import reduce
# nums = [1, 2, 3, 4]
# result = reduce(lambda a, b: a + b, nums, 10)
# print("Result:", result)
# Output:Result: 20


# 🔎How the Initial Value Affects the Process
# Syntax:reduce(function, iterable, initial_value)
# Here:
# Function → lambda a, b: a + b
# Iterable → [1, 2, 3, 4]
# Initial value → 10

# 🧠 Step-by-Step Execution
# Since we gave 10 as the initial value, reduction starts like this:
# Step 1: 10 + 1 = 11
# Step 2: 11 + 2 = 13
# Step 3: 13 + 3 = 16
# Step 4: 16 + 4 = 20
# Final result = 20

# ❓ What If No Initial Value Is Given?
# reduce(lambda a, b: a + b, nums)
# Then it works like:
# Step 1: 1 + 2 = 3
# Step 2: 3 + 3 = 6
# Step 3: 6 + 4 = 10
# Result = 10

# 📌 Key Difference
# | Case                    | Starting Point     | Final Result |
# | ----------------------- | ------------------ | ------------ |
# | Without initial value   | First two elements | 10           |
# | With initial value (10) | Starts from 10     | 20           |

# ⭐ Important Notes
# The initial value acts as the starting accumulator value.
# It is added before processing the list.
# If the list is empty and initial value is given → it returns the initial value.
# If the list is empty and no initial value → ❌ Error.

# ✅ Final Understanding
# The initial value:
# Changes the starting point of reduction
# Can change the final result
# Makes the function safer for empty lists
```

```python
# 5. Consider the code below:
# nums = [[1, 2], [3, 4], [5, 6]] result = list(map(lambda x: x.append(10), nums))
# print("Result:", result) print("Nums:", nums)
# Questions
# • What will be the output of result?
# • What will be the output of nums?
# • Why does map() behave this way with list.append()?
# • How can you modify the lambda so that nums is not changed?

# nums = [[1, 2], [3, 4], [5, 6]]
# result = list(map(lambda x: x.append(10), nums))
# print("Result:", result)
# print("Nums:", nums)

# ✅ Output:
# Result: [None, None, None]
# Nums: [[1, 2, 10], [3, 4, 10], [5, 6, 10]]


# ✅ 1️⃣ What will be the output of result?
# Result: [None, None, None]
# 🔎 Why?
# list.append() modifies the list in place
# It does NOT return anything
# In Python, if a function returns nothing → it returns None
# So each time x.append(10) runs, it returns None.
# That’s why result becomes:[None, None, None]


# ✅ 2️⃣ What will be the output of nums?
# Nums: [[1, 2, 10], [3, 4, 10], [5, 6, 10]]
# 🔎 Why?
# append() changes the original inner lists directly.
# map() is passing each inner list to append(), and it modifies them.


# ✅ 3️⃣ Why does map() behave this way with list.append()?
# map() collects the return value of the function.
# append() returns None.
# But it modifies the original list in memory.
# So:
# | Action           | What Happens  |
# | ---------------- | ------------- |
# | `append()`       | Modifies list |
# | Return value     | `None`        |
# | `map()` collects | `None` values |
# So map() is not the problem — append() returns None.


# ✅ 4️⃣ How to Modify Lambda So nums Is NOT Changed?
# Instead of modifying the original list, create a new list:
# ✔ Correct Way:
# nums = [[1, 2], [3, 4], [5, 6]]
# result = list(map(lambda x: x + [10], nums))
# print("Result:", result)
# print("Nums:", nums)


# ✅ Now Output Will Be:
# Result: [[1, 2, 10], [3, 4, 10], [5, 6, 10]]
# Nums: [[1, 2], [3, 4], [5, 6]]
# 🔎 Why This Works
# x + [10]:
# Creates a new list
# Does NOT modify original list
# Returns the new list
# So map() collects the correct values


# 🧠 Important Concept
# | Method     | Modifies Original? | Returns Value?     |
# | ---------- | ------------------ | ------------------ |
# | `append()` | ✅ Yes              | ❌ Returns None     |
# | `x + [10]` | ❌ No               | ✅ Returns new list |


# ⭐ Interview Tip
# Avoid using append() inside map() because:
# It returns None
# It causes side effects
# It makes code harder to understand
```
